```python
import json
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
```

```python
# Télécharger les ressources nécessaires de NLTK
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

Tokenization de data.json, dictionnaire ayant pour clé l'id d'un document wikipedia et pour value associée la segmentation de cet
article en passages de 6 phrases

```python
def load_data(filepath):
    with open(filepath, 'r') as file:
        data = json.load(file)
    return data

filepath = './data.json'
wiki_docs = load_data(filepath)
```

```python
print(len(wiki_docs['672']))
print(wiki_docs['672'][0])   
```

```python
# Initialisation de la liste de stop words et du lemmatiseur
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    # Tokenisation
    tokens = word_tokenize(text)
    # Suppression des mots vides et nettoyage des caractères non alphabétiques
    tokens = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]
    # Lemmatisation
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return tokens

# Parcours de vos documents et application de la fonction de prétraitement
for key, article in wiki_docs.items():
    for i in range(len(article)):
        wiki_docs[key][i] = preprocess_text(wiki_docs[key][i])
```

```python
print(len(wiki_docs['672'][0]))
print(wiki_docs['672'][0]) 
```

```python
with open('data_tokenized.json', 'w') as f:
    json.dump(wiki_docs, f)  
```

On fait la même pour le dictionnaire qui a pour clé l'id des documents, et pour valeur associée à cette clef un dictionnaire listant tous les passages du document (dont les vrais labels, ie les "relevant passages") (c'est le dico qui était donné de base)                                          

```python
filepath = 'WikiPassageQA/document_passages.json'
doc_passages = load_data(filepath)
```

```python
print(doc_passages.keys())
print(doc_passages["672"])
```

```python
for doc_id, passages_dict in doc_passages.items():
    for passage_number, passage in passages_dict.items():
        doc_passages[doc_id][passage_number] = preprocess_text(doc_passages[doc_id][passage_number])
```
